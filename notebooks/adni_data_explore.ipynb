## Retrieve data for ADNI1

```python
adni1_mrqc = pd.read_csv("MRIQUALITY_ADNI1.csv")
adni1_mrqc_3t = adni1_mrqc[adni1_mrqc["FLDSTRENG"]==3]
adni1_mrqc_3t_T1 = adni1_mrqc_3t[adni1_mrqc_3t["SERIESDESC"].str.contains("RAGE")]
print(len(adni1_mrqc_3t_T1))
adni1_mrqc_3t_T1.head()
```

```python
len(adni1_mrqc_3t_T1[['PTID', "VISCODE"]].drop_duplicates())
```

```python
adni1_mrqc_3t_T1['motion_in_comment']=(adni1_mrqc_3t_T1["COMMENTS"].str.contains("ringing") | adni1_mrqc_3t_T1["COMMENTS"].str.contains("motion") )
adni1_mrqc_3t_T1['IPMOTION'].unique()
```

Now that we have every 3T T1w volumes in ADNI 1, we want to know which one is kept for each visit (Reminder : ADNI 1/Go/2 take two T1 volumes for each patient visit). For that, we will use the RECNO (record number) property of the full meta parameter list

```python
adni1_meta = adni_mri_meta[adni_mri_meta['PHASE']=='ADNI1']
len(adni1_meta)
```

```python
len(adni1_meta[['PTID', "VISCODE"]].drop_duplicates())
```

```python
mriqc_unique =set(map(tuple, adni1_mrqc_3t_T1[['PTID', "VISCODE"]].drop_duplicates().to_numpy()))
meta_unique =set(map(tuple, adni1_meta[['PTID', "VISCODE"]].drop_duplicates().to_numpy()))
len(meta_unique.difference(mriqc_unique)), adni1_meta['HAS_QC_ERROR'].isna().count()
```

## Retrieve Data for ADNI2/GO

```python
adni2_qc = pd.read_csv("MAYOADIRL_MRI_IMAGEQC_ADNI2.csv")
adni2_qc_t1 = adni2_qc[adni2_qc['series_type']=="T1"]
adni2_qc_t1 = adni2_qc_t1[adni2_qc_t1['field_strength']==3]

len(adni2_qc_t1), adni2_qc_t1.columns, len(adni2_qc_t1['series_description'].unique()), adni2_qc_t1
```

```python
len(adni2_qc_t1[['PTID', "VISCODE"]].drop_duplicates())
```

```python
adni2_meta =  adni_mri_meta[(adni_mri_meta['PHASE']=='ADNI2') | (adni_mri_meta['PHASE']=='ADNIGO')]
len(adni2_meta)
```

```python
adni2_qc_t1['IMAGEUID']=adni2_qc_t1['loni_image'].str.removeprefix("I")
adni2_qc_t1_no_na=adni2_qc_t1[adni2_qc_t1['IMAGEUID'].notna()]
adni2_qc_t1_no_na['IMAGEUID'] = adni2_qc_t1_no_na['IMAGEUID'].astype(np.int32)
rm_na_list = mri_list[mri_list['IMAGEUID'].notna()].set_index("IMAGEUID")

adni2_qc_with_subject = adni2_qc_t1_no_na.join(rm_na_list[["SUBJECT", "SCANDATE", "VISIT"]], "IMAGEUID", "inner", rsuffix="_right")

adni2_meta['date']=adni2_meta['EXAMDATE'].str.replace("-","").fillna(-1).astype(np.int32)
adni2_meta_ptid = adni2_meta.set_index(["PTID", "date"])

adni2_all =adni2_qc_with_subject.join(adni2_meta_ptid[['MOTION', 'RECNO']], ["SUBJECT", "series_date"], "inner")
adni2_all
```

```python
adni2_clean = adni2_all[adni2_all['MOTION']==0]
adni2_clean
```
